```python
# 1. Convertimos detectando automÃ¡ticamente ambos formatos
df_precio_vivienda['Mes'] = pd.to_datetime(df_precio_vivienda['Mes'], dayfirst=True, format='mixed')

# 2. Ahora que todos son fechas reales, unificamos visualmente a YYYY-MM
df_precio_vivienda['Mes'] = df_precio_vivienda['Mes'].dt.strftime('%Y-%m')
```

```python
# Renombrar la columna de fecha a un nombre estÃ¡ndar (ej. 'fecha')
# Sustituye 'Nombre_Actual' por el nombre que tenga ahora esa columna
df_precio_vivienda =df_precio_vivienda.rename(columns={'Mes': 'DateTime'})

# Convertir esa columna a formato datetime
df_precio_vivienda['DateTime'] = pd.to_datetime(df_precio_vivienda['DateTime'])

# Crear la columna 'Anio' extrayendo el aÃ±o del datetime
df_precio_vivienda['Anio'] = df_precio_vivienda['DateTime'].dt.year

# Sobrescribir el CSV original con los cambios
# Usamos index=False para que no te cree una columna de nÃºmeros extra
df_precio_vivienda.to_csv('precio_vivienda_final.csv', index=False)

print("Â¡Archivo actualizado correctamente con la columna Anio!")
```
