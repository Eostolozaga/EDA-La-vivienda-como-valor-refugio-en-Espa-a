```python
import yfinance as yf
import seaborn as sns
import matplotlib as plt
import pandas as pd

```

```python
df_oro_dolares = yf.download("GC=F", period="10y")

df_oro_dolares.columns = df_oro_dolares.columns.get_level_values(0)

df_oro_dolares = df_oro_dolares.reset_index()

df_oro_dolares.head()
```

```python
# 1. Descarga de datos del oro (USD) y del tipo de cambio (EUR/USD)
df_oro_dolares = yf.download("GC=F", period="10y")
df_eur_usd = yf.download("EURUSD=X", period="10y")

# 2. Limpiar niveles de columnas
df_oro_dolares.columns = df_oro_dolares.columns.get_level_values(0)
df_eur_usd.columns = df_eur_usd.columns.get_level_values(0)

# 3. Crear el nuevo DataFrame en Euros
# Dividimos el precio en dólares por el valor del tipo de cambio
df_oro_euros = df_oro_dolares.copy()
df_oro_euros['Close'] = df_oro_dolares['Close'] / df_eur_usd['Close']

# 4. Resetear índice si lo necesitas
df_oro_euros = df_oro_euros.reset_index()

df_oro_euros.head()
```

```python
# 1. Aseguramos que la fecha sea datetime (yfinance suele traerla en el índice o como 'Date')
# Si la fecha es el índice, usa: df_oro = df_oro.reset_index()
df_oro_euros['Date'] = pd.to_datetime(df_oro_euros['Date'])

# 2. Creamos la columna con el formato 'YYYY-MM' (la que usaremos para el merge)
df_oro_euros['Mes'] = df_oro_euros['Date'].dt.strftime('%Y-%m')

# 3. Calculamos la MEDIA MENSUAL del precio (ej. de la columna 'Close')
# Agrupamos por la nueva columna 'Mes' y calculamos la media
df_oro_mensual = df_oro_euros.groupby('Mes')['Close'].mean().reset_index()

# 4. Renombramos la columna para que se entienda que es una media
df_oro_mensual = df_oro_mensual.rename(columns={'Close': 'oro_media_mensual'})

# 5. Ver el resultado y exportar
df_oro_mensual.head()

```

```python
df_oro_mensual.to_csv('oro_mensual_limpio.csv', index=False)
```

```python
# Renombrar la columna de fecha a un nombre estándar (ej. 'fecha')
# Sustituye 'Nombre_Actual' por el nombre que tenga ahora esa columna
df_oro_mensual =df_oro_mensual.rename(columns={'Mes': 'DateTime'})

# Convertir esa columna a formato datetime
df_oro_mensual['DateTime'] = pd.to_datetime(df_oro_mensual['DateTime'])

# Crear la columna 'Anio' extrayendo el año del datetime
df_oro_mensual['Anio'] = df_oro_mensual['DateTime'].dt.year

# Sobrescribir el CSV original con los cambios
# Usamos index=False para que no te cree una columna de números extra
df_oro_mensual.to_csv('oro_final.csv', index=False)

print("¡Archivo actualizado correctamente con la columna Anio!")
```
