```python
# Renombrar la columna de fecha a un nombre estándar (ej. 'fecha')
# Sustituye 'Nombre_Actual' por el nombre que tenga ahora esa columna
poder_adquisitivo =poder_adquisitivo.rename(columns={'Periodo': 'DateTime'})

# Convertir esa columna a formato datetime
poder_adquisitivo['DateTime'] = pd.to_datetime(poder_adquisitivo['DateTime'])

# Crear la columna 'Anio' extrayendo el año del datetime
poder_adquisitivo['Anio'] = poder_adquisitivo['DateTime'].dt.year

# Sobrescribir el CSV original con los cambios
# Usamos index=False para que no te cree una columna de números extra
poder_adquisitivo.to_csv('poder_adquisitivo_final.csv', index=False)

print("¡Archivo actualizado correctamente con la columna Anio!")
```

```python
# 1. Extraer el año real de la columna DateTime (los últimos 4 caracteres)
poder_adquisitivo['Anio'] = poder_adquisitivo['DateTime'].astype(str).str[-4:].astype(int)

# 2. Limpiar la columna Total (Ej: '1.767,6' -> 1767.6)
# Quitamos el punto de miles y cambiamos la coma por punto
poder_adquisitivo['Total_Num'] = (poder_adquisitivo['Total']
                       .str.replace('.', '', regex=False)
                       .str.replace(',', '.', regex=False)
                       .astype(float))

# 3. Anualizar salario (Mensual * 12)
poder_adquisitivo['Salario_Anual_Real'] = poder_adquisitivo['Total_Num'] * 12
```

```python
poder_adquisitivo.to_csv('poder_adquisitivo_final.csv', index=False)
```
