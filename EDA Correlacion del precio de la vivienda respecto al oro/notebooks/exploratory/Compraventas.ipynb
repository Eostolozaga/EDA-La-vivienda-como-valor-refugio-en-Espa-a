```python
import pandas as pd
compraventas = pd.read_csv(r'C:\Users\Eneko\Desktop\Ejercicios Data\Ejercicios\Proyecto EDA\EDA Correlacion del precio de la vivienda respecto al oro\CSV\Compraventas.csv', sep=';', encoding='utf-8-sig')
fuente_6 = "https://www.ine.es/consul/serie.do?d=true&s=ETDP377&c=2&"
```

```python
compraventas.head()
```

```python
compraventas['Anio'] = compraventas['PERIODO'].str[:4].astype(int)

# Convertimos VALOR a número (el INE usa el punto como separador de MILES aquí)
compraventas['VALOR_Num'] = pd.to_numeric(
    compraventas['VALOR'].astype(str).str.replace('.', '', regex=False), 
    errors='coerce'
)

# 3. Agrupamos por Año (SUMA) 
# Al ser volumen, sumamos todos los meses para tener el total de ventas del año
compraventas_anual = compraventas.groupby('Anio')['VALOR_Num'].sum().reset_index()
compraventas_anual.columns = ['Anio', 'Total_Ventas']
```

```python
compraventas_anual.head()
```

```python
# 1. Aseguramos que los datos estén ordenados cronológicamente antes de calcular
compraventas_anual = compraventas_anual.sort_values('Anio')

# 2. Añadimos la columna de Variación Anual (%)
# (Valor actual - Valor anterior) / Valor anterior * 100
compraventas_anual['Var_Ventas_Pct'] = compraventas_anual['Total_Ventas'].pct_change() * 100
```
