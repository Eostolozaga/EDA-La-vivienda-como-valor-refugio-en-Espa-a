```python
# Función para corregir el valor
def corregir_escalado(valor):
    # Si el valor ya es >= 600 o <= 0, no lo tocamos (o ajusta según necesites)
    if valor >= 600 or valor == 0:
        return valor
    
    # Multiplicamos por 10 mientras sea menor a 100
    while valor < 100:
        valor *= 10
        
    return valor

# Aplicarlo a tu columna
df_tipos_interes['Préstamos y créditos (TAE)'] = df_tipos_interes['Préstamos y créditos (TAE)'].apply(corregir_escalado)
```

```python
# Convertimos la columna a numérica (importante para que la función no falle)
df_tipos_interes['Préstamos y créditos (TAE)'] = pd.to_numeric(
    df_tipos_interes['Préstamos y créditos (TAE)'], 
    errors='coerce'
)

# Opcional: Rellenar nulos con 0 para que la función pueda procesarlos
df_tipos_interes['Préstamos y créditos (TAE)'] = df_tipos_interes['Préstamos y créditos (TAE)'].fillna(0)

df_tipos_interes['Préstamos y créditos (TAE)'] = df_tipos_interes['Préstamos y créditos (TAE)'].apply(corregir_escalado)

todo_ok = (df_tipos_interes['Préstamos y créditos (TAE)'] >= 100).all()
print(f"¿Todos superan 100?: {todo_ok}")
```

```python
# 1. Convertimos a fecha real especificando el formato exacto (día/mes/año de 2 dígitos)
df_tipos_interes['DateTime'] = pd.to_datetime(df_tipos_interes['DateTime'], format='%d/%m/%y')

# 2. Pasamos al formato unificado YYYY-MM
df_tipos_interes['DateTime'] = df_tipos_interes['DateTime'].dt.strftime('%Y-%m')
```
