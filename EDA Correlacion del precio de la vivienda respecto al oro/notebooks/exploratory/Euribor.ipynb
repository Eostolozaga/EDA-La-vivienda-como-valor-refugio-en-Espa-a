```python
import pandas as pd
# guarda en variables los datasets y su fuente
df_euribor = pd.DataFrame(pd.read_csv(r"C:\Users\Eneko\Desktop\Ejercicios Data\Ejercicios\Proyecto EDA\EDA Correlacion del precio de la vivienda respecto al oro\CSV\Euribor 2015-2025.csv"))
fuente_1 = "https://www.epdata.es/datos/euribor-hoy-evolucion-historica-euribor-mensual-estadisticas-datos-hipotecas-variables/502"
```

```python
# Diccionario de traducción
meses_es_en = {
    'ene': 'Jan', 'feb': 'Feb', 'mar': 'Mar', 'abr': 'Apr',
    'may': 'May', 'jun': 'Jun', 'jul': 'Jul', 'ago': 'Aug',
    'sep': 'Sep', 'oct': 'Oct', 'nov': 'Nov', 'dic': 'Dec'
}

# 1. Aseguramos que el mes esté en inglés (usando el diccionario que ya tienes)
df_euribor['mes_en'] = df_euribor['Fecha'].str[:3].str.lower().map(meses_es_en)

# 2. Extraemos el año saltando el posible guion original
# Si tus datos son 'ene-99', el año empieza en la posición 4. 
# Si son 'ene99', empieza en la 3.
# Esta línea busca solo los números:
df_euribor['anio_num'] = df_euribor['Fecha'].str.extract('(\d+)')

# 3. Concatenamos SIN guion extra y usamos el formato '%b%y'
df_euribor['fecha_dt'] = pd.to_datetime(df_euribor['mes_en'] + df_euribor['anio_num'], format='%b%y')
```

```python
# Filtrar entre el 1 de enero de 2015 y el 31 de diciembre de 2025
mask = (df_euribor['fecha_dt'] >= '2015-01-01') & (df_euribor['fecha_dt'] <= '2025-12-31')
df_euribor_filtrado = df_euribor.loc[mask]
df_euribor_filtrado.head()
```

```python
df_euribor['fecha_dt'] = pd.to_datetime(df_euribor['fecha_dt'])

# Quitar los días (Convertir a Periodo de Mes y luego a Texto YYYY-MM)
df_euribor['fecha_dt'] = df_euribor['fecha_dt'].dt.strftime('%Y-%m')
```

```python
# Renombrar la columna de fecha a un nombre estándar (ej. 'fecha')
# Sustituye 'Nombre_Actual' por el nombre que tenga ahora esa columna
df_euribor =df_euribor.rename(columns={'fecha_dt': 'DateTime'})

# Convertir esa columna a formato datetime
df_euribor['DateTime'] = pd.to_datetime(df_euribor['DateTime'])

# Crear la columna 'Anio' extrayendo el año del datetime
df_euribor['Anio'] = df_euribor['DateTime'].dt.year

# Sobrescribir el CSV original con los cambios
# Usamos index=False para que no te cree una columna de números extra
df_euribor.to_csv('euribor_final.csv', index=False)

print("¡Archivo actualizado correctamente con la columna Anio!")
```
